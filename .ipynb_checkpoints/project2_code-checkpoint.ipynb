**The tweets related to Donald trump and Joe Biden are extracted from Twitter database by search of key hashtags. Thus, the two datasets only represent the mentioning of a candidate, but not necessarily their political stance. A sizable portion of the tweet is in other languages like Spanish and German, and written by non-USA residents. ##### Stop words and lemmatization

```python
# Identify part of speech
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)

# Function to clean tweet
def clean_tweet(tweet, lemmatize = 'nltk'):
    """
    Inputs:
    tweet - a string representing the text we need to clean
    lemmatize - one of two possible values {spacy, nltk} 
      two lemmatization methods
      with our dataset, we got the best results with nltk
      but Spacy also did a good job, hence you might 
      try both and compare results for your own data
    
    Output: 
    tokenized - the cleaned text, tokenized (a list of string words)
    """

    tweet = tweet.lower() # lowercase
    tweet = re.sub(r"http\S+|www\S+|https\S+", '', tweet, flags=re.MULTILINE) # remove urls
    tweet = re.sub(r'\@\w+|\#','', tweet) # remove mentions of other usernames and the hashtag character
    tweet = remove_stopwords(tweet) # remove stopwords with Gensim

    if (lemmatize == 'spacy'):
        # Initialize spacy 'en' model, keeping only tagger component needed for lemmatization
        nlp = spacy.load('en', disable=['parser', 'ner'])
        doc = nlp(tweet)
        tokenized = [token.lemma_ for token in doc if token.lemma_ != '-PRON-']
    elif (lemmatize == 'nltk'): 
        '''
        lemmatization works best when WordNetLemmatizer receives both the text and the part of speech of each word
        the code below assigns POS (part of speech) tag on a per word basis (it does not infer POS from contenxt / sentence), which might not be optimal
        '''
        lemmatizer = WordNetLemmatizer()
        tokenized = [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in nltk.word_tokenize(tweet)]
    
    # remove left over stop words with nltk
    tokenized = [token for token in tokenized if token not in stopwords.words("english")] 

    # remove non-alpha characters and keep the words of length >2 only
    tokenized = [token for token in tokenized if token.isalpha() and len(token)>2]

    return tokenized

# clean tweet string
def combine_tokens(tokenized): 
    non_tokenized = ' '.join([w for w in tokenized])
    return non_tokenized
```

```python
# Import class
from transformers import pipeline
from transformers import AutoModelForSequenceClassification
from transformers import TFAutoModelForSequenceClassification
from transformers import AutoTokenizer
import numpy as np
from scipy.special import softmax
import csv
import urllib.request

# Load RoBERTa's sentiment model

tokenizer = AutoTokenizer.from_pretrained('cardiffnlp/twitter-roberta-base-sentiment')

model = AutoModelForSequenceClassification.from_pretrained('cardiffnlp/twitter-roberta-base-sentiment')

# Use the model on each tweet to perform sentiment analysis and seperate scores for positve, neutral and negative

# download label mapping
mapping_link = f"https://raw.githubusercontent.com/cardiffnlp/tweeteval/main/datasets/sentiment/mapping.txt"
with urllib.request.urlopen(mapping_link) as f:
    html = f.read().decode('utf-8').split("\n")
    csvreader = csv.reader(html, delimiter='\t')
labels = [row[1] for row in csvreader if len(row) > 1]
print(labels)
```

```python
df = pd.read_csv('data/tweets/clean_stance_topic.csv')
```

```python
dom_topic = df[df.perc_contribution >0.6]
```

```python
# Generate each tweet's sentiment percentages for the first ten tweets
sentiment_perc = []
sentiment_label = []
label_perc = []
for tweet in dom_topic['clean_tweet'][0:10].tolist():
    encoded_tweet = tokenizer(tweet, return_tensors='pt')
    output = model(**encoded_tweet)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    roberta_senti = dict(zip(labels, list(scores[0:3])))
    max_key = max(roberta_senti, key=roberta_senti.get)
    max_value = max(roberta_senti.values())
    sentiment_perc.append(roberta_senti)
    sentiment_label.append(max_key)
    label_perc.append(max_value)
    print(roberta_senti)
    print(max_key)
    print(max_value)
```

##### Convert the most popular topics dataframe tweets to sentiment

```python
# Generate each tweet's sentiment percentages for tweets
sentiment_perc = []
sentiment_label = []
label_perc = []

start_time = time.time()

for tweet in df2['clean_tweet'].tolist():
    encoded_tweet = tokenizer(tweet, return_tensors='pt')
    output = model(**encoded_tweet)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    roberta_senti = dict(zip(labels, list(scores[0:3])))
    max_key = max(roberta_senti, key=roberta_senti.get)
    max_value = max(roberta_senti.values())
    sentiment_perc.append(roberta_senti)
    sentiment_label.append(max_key)
    label_perc.append(max_value)

stop_time = time.time()

print(f'It took {np.around((time.time() - start_time),decimals=1)} seconds')
```
